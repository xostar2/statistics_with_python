```python
#outliears
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Define our Dataset
dataset = [
    45, 48, 50, 52, 49, 51, 47, 53, 54, 46,
    50, 49, 52, 48, 51, 47, 53, 54, 46, 50,
    49, 52, 48, 51, 47,
    120,   # high outlier
    130,   # high outlier
    5,     # low outlier
    8      # low outlier
]
```

```python
outliear=[]
#z_score computation
def detect_ouliers(data):
    threshhold=3 #threshold is third of standard deviation
    mean=np.mean(data)
    std=np.std(data)

    for i in data:
        z_score=(i-mean)/std
        if abs(z_score)>threshhold:
            outliear.append(i)
    return outliear
```

```python
plt.hist(dataset)
```

```python
outliear_list=detect_ouliers(dataset)
```
