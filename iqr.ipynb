```python
#outliears
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Define our Dataset
dataset = [
    45, 48, 50, 52, 49, 51, 47, 53, 54, 46,
    50, 49, 52, 48, 51, 47, 53, 54, 46, 50,
    49, 52, 48, 51, 47,
    120,   # high outlier
    130,   # high outlier
    5,     # low outlier
    8      # low outlier
]
```

```python
outliear=[]
#z_score computation
def detect_ouliers(data):
    threshhold=3 #threshold is third of standard deviation
    mean=np.mean(data)
    std=np.std(data)

    for i in data:
        z_score=(i-mean)/std
        if abs(z_score)>threshhold:
            outliear.append(i)
    return outliear
```

```python
plt.hist(dataset)
```

```python
outliear_list=detect_ouliers(dataset)
```

```python
print(outliear_list)
```

```python
# #IQR

# 1. IQR is the difference between the 75th percentile and the 25th percentile of a distribution.
# 2. The IQR is a measure of the dispersion of a distribution.
# 3. The IQR is a measure of the spread of a distribution.
# 4.sort the data
# 5.IQR=Q3-Q1
# 6.find the lower fence(q1.(1.5*iqr))
# 7.find the upper fence(q3.(1.5*iqr))
```

```python
dataset=sorted(dataset)
```

```python
q1,q3=np.percentile(dataset,[25,75])
iqr=q3-q1
print(q1,q3,iqr)


```

```python
#find lower_fence and higher_fence
lower_fence=q1-(1.5*iqr)
print(lower_fence)

higher_fence=q3+(1.5*iqr)
print(higher_fence)

#find outliers
```
